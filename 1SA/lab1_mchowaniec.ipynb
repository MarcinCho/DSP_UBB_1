```python
import numpy as np
import matplotlib.pyplot as plt

X_mu = np.array([6, 5, 4, 3, 4, 5, 0, 0, 0, 0])
N = len(X_mu)  # N = 10

n_indices = np.arange(N)
K = np.outer(n_indices, n_indices)

W = np.exp(1j * 2 * np.pi / N * K)

x_k = (1 / N) * W @ X_mu

print(f"Block Length N: {N}")
print("\nMatrix K (10x10):")
print(K)

print("\nMatrix W (First 3 rows, rounded):")
print(W)

print("\nSynthesized Signal x[k] (Complex values):")
print(np.round(x_k, 3))
```

```python
# --- Visualization ---
# Plotting the real and imaginary parts of the synthesized signal
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.stem(n_indices, np.real(x_k), basefmt=" ", label='Real Part')
plt.title(r'Synthesized Signal $x[k]$ (Real Part)')
plt.xlabel(r'Time Index $k$')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.stem(n_indices, np.imag(x_k), linefmt='C1-', markerfmt='C1o', basefmt=" ", label='Imaginary Part')
plt.title(r'Synthesized Signal $x[k]$ (Imaginary Part)')
plt.xlabel(r'Time Index $k$')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# --- Plotting ---
plt.figure(figsize=(10, 6))

k_idx = np.arange(N)
mu_idx = np.arange(N)

# Real Part - Connected lines with markers
plt.plot(k_idx, np.real(x_k), 'o-', label='Real Part', color='blue')

# Imaginary Part - Connected lines with markers
plt.plot(k_idx, np.imag(x_k), 'o-', label='Imaginary Part', color='orange')

plt.title(r'Synthesized Signal $x[k]$ (Real & Imaginary)')
plt.xlabel(r'Time Index $k$')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
